```python
#stopwords
new_sentence = [text for text in sentence if not text in textfile]
def handle_num(new_sentence):
    for i in range(len(new_sentence)):
        if new_sentence[i][-1] >= "1" and  new_sentence[i][-1] <= "4":
            new_sentence[i] = new_sentence[i][:-1]
        
    return new_sentence
def handle_empty(new_sentence):
    while('' in new_sentence):
        new_sentence.remove('')
    return new_sentence

def handle_comma(new_sentence):
    print(new_sentence)
    for i in range(len(new_sentence)):
        
        if new_sentence[i][-1] == ",":
            new_sentence[i] = new_sentence[i][:-1]
    return new_sentence
```

```python
vowel = ['a','e','i','o','u']
# test_sentence = ['gaps','gas','stress']

def sses_ss(new_sentence):
    for i in range(len(new_sentence)):
        if new_sentence[i].endswith('sses'):
            new_sentence[i] = new_sentence[i].replace('sses','ss')
    return new_sentence

#To check preceding word part contains a vowel
def check_contain(j):
    cnt = 0
    if j in vowel:
        cnt += 1
        return cnt
    else:
        return 0
def suffix_rm_s(new_sentence):
    for i in range(len(new_sentence)):
        #To check vowel not immediately before the s
        #If suffix is us or ss do nothing
        if (new_sentence[i][-2:] != 'us' and new_sentence[i][-2:] != 'ss') and new_sentence[i][-1] == 's' and new_sentence[i][-2] not in vowel and len(new_sentence[i])>2:
            for j in new_sentence[i]:
                if check_contain(j):
                    #Delete s for the last character
                    new_sentence[i] = new_sentence[i][:-1]
    return new_sentence
```

```python
# new_sentence = ['ties','cries']
def ied_to_ie(new_sentence):
    for i in range(len(new_sentence)):
        # otherwise by ie
        if len(new_sentence[i][:-3])==1 and (new_sentence[i][-3:] == 'ies' or new_sentence[i][-3:] == 'ied'):
            new_sentence[i] = new_sentence[i][:-1]

        #Replace ied or ies by i if preceded by more than one letter, 
        elif len(new_sentence[i][:-3])>1 and (new_sentence[i][-3:] == 'ies' or new_sentence[i][-3:] == 'ied'):
            new_sentence[i] = new_sentence[i][:-2]
            
    return new_sentence
    
```

```python
# test_sentence3 = ['agreed','feed']

def eed_to_ee(new_sentence):
    for i in range((len(new_sentence))):
        if new_sentence[i][0] in vowel and new_sentence[i][1] not in vowel and (new_sentence[i][-3:] == 'eed' or new_sentence[i][-5:] == 'eedly'):
            new_sentence[i] = new_sentence[i][:-1]
    return new_sentence            
```

```python
test_sentence4 = ['fished','pirating','falling','dripping','hoping']
check_ll_ss_zz = ['ll','ss','zz']
consonant = ['b','c','d','f','g,','h','j','k','l','m','n','p','q','r','s','t','v','w','y','x','z']
vowel = ['a','e','i','o','u']

def step_1b_2(i):
    if new_sentence[i][-2:] == 'at' or new_sentence[i][-2:] == 'bl' or new_sentence[i][-2:] == 'iz':
        new_sentence[i] = new_sentence[i] + 'e'
    elif len(set(new_sentence[i][-2:]))==1 and new_sentence[i][-2:] not in check_ll_ss_zz:
        new_sentence[i] = new_sentence[i][:-1]
    
    if short_word(new_sentence[i]): new_sentence[i] = new_sentence[i] + 'e'
        
    return new_sentence[i]

#short
def short_word(new_sentence):    
    for i in range(len(new_sentence)):
        
        if new_sentence[0] in vowel and new_sentence[-1] in vowel: 
            for c in set(new_sentence[1:-1]):
                if c in consonant:
                    pass
                elif c in vowel:
                    return False
            return True    
        
        if new_sentence[i].startswith(tuple(vowel)):
            for v in vowel:
                if new_sentence[i][1:].find(v) == -1:
                    pass
                else:
                    return False
        
        if new_sentence[i].startswith(tuple(consonant))==1 and new_sentence[i].endswith(tuple(consonant))==1:
            for v in new_sentence[1:-1]:
                
                if v in vowel:
                    pass
                elif v in consonant:
                    return False
            return True
                    
    
def delete_ed_ing(new_sentence):
    for i in range(len(new_sentence)):
        vowel_cnt = 0
        for j in new_sentence[i]:
            if j in vowel:
                vowel_cnt+=1
        if vowel_cnt>1:
            
            if new_sentence[i][-2:] == 'ed':
                new_sentence[i] = new_sentence[i].replace('ed','')
                step_1b_2(i)
                    
            elif new_sentence[i][-4:] == 'edly':
                new_sentence[i] = new_sentence[i].replace('edly','')
                step_1b_2(i)
                    
            elif new_sentence[i][-3:] == 'ing':
                new_sentence[i] = new_sentence[i].replace('ing','')
                step_1b_2(i)
                    
            elif new_sentence[i][-5:] == 'ingly':
                new_sentence[i] = new_sentence[i].replace('ingly','')
                step_1b_2(i)
            
            else:
                step_1b_2(i)
                    
    return new_sentence

```

```python
def all_func_gathered(new_sentence):
    new_sentence = sses_ss(new_sentence)
    new_sentence = suffix_rm_s(new_sentence)
    new_sentence = ied_to_ie(new_sentence)
    new_sentence = handle_num(new_sentence)
    new_sentence = handle_empty(new_sentence)
    new_sentence = handle_comma(new_sentence)
    new_sentence = eed_to_ee(new_sentence)
    new_sentence = delete_ed_ing(new_sentence)
    return new_sentence
all_func_gathered(new_sentence)
```
